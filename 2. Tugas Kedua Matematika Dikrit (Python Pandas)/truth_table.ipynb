# Jawaban No 1

## a) p → (¬q ∨ r)

```python
tabel = []

for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = not q
            prop2 = prop1 or r
            prop3 = not p or prop2

            tabel.append(pd.DataFrame({
                'P': [p],
                ' ': ['  '],
                'Q': [q],
                '  ': ['  '],
                'R': [r],
                '   ': ['  '],
                '~q': [prop1],
                '    ': ['  '],
                '~q || r': [prop2],
                '     ': ['  '],
                'p -> (~q || r)': [prop3],
            }))

tabel_soal_1_a = pd.concat(tabel, ignore_index=True)
```

## b) ¬p → (q → r)

```python
tabel = []

for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = not p
            prop2 = not q or r
            prop3 = not prop1 or prop2

            tabel.append(pd.DataFrame({
                'P': [p],
                ' ': [' '],
                'Q': [q],
                '  ': [' '],
                'R': [r],
                '   ': [' '],
                '~p': [prop1],
                '    ': [' '],
                'q -> r': [prop2],
                '     ': [' '],
                '~p -> (q -> r)': [prop3],
            }))

tabel_soal_1_b = pd.concat(tabel, ignore_index=True)
```

## c) (p → q) ∨ (¬p → r)

```python
tabel = []

for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = not p
            prop2 = not p or q
            prop3 = not prop1 or r
            prop4 = prop2 or prop3

            tabel.append(pd.DataFrame({
                'P': [p],
                ' ': [' '],
                'Q': [q],
                '  ': [' '],
                'R': [r],
                '   ': [' '],
                '¬p': [prop1],
                '    ': [' '],
                'p -> q': [prop2],
                '      ': [' '],
                '¬p -> r': [prop3],
                '       ': [' '],
                '(p -> q) || (¬p -> r)': [prop4],
            }))

tabel_soal_1_c = pd.concat(tabel, ignore_index=True)


```

## d) (p → q) ∧ (¬p → r)

```python
tabel = []

for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = not p
            prop2 = not p or q
            prop3 = not prop1 or r
            prop4 = prop2 and prop3

            tabel.append(pd.DataFrame({
                'P': [p],
                ' ': [' '],
                'Q': [q],
                '  ': [' '],
                'R': [r],
                '   ': [' '],
                '¬p': [prop1],
                '    ': [' '],
                'p -> q': [prop2],
                '      ': [' '],
                '¬p -> r': [prop3],
                '       ': [' '],
                '(p -> q) && (¬p -> r)': [prop4],
            }))

tabel_soal_1_d = pd.concat(tabel, ignore_index=True)


```

## e) (p ↔ q) ∨ (¬q ↔ r)

```python
tabel = []

for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = not q
            prop2 = p == q
            prop3 = prop1 == r
            prop4 = prop2 or prop3

            tabel.append(pd.DataFrame({
                'P': [p],
                ' ': [' '],
                'Q': [q],
                '  ': [' '],
                'R': [r],
                '   ': [' '],
                '~q': [prop1],
                '    ': [' '],
                'p <--> q': [prop2],
                '      ': [' '],
                '~q <--> r': [prop3],
                '       ': [' '],
                '(p <--> q) || (~q <--> r)': [prop4],
            }))

tabel_soal_1_e = pd.concat(tabel, ignore_index=True)

```

## f) (¬p ↔ ¬q) ↔ (q ↔ r)

```python
tabel = []

for p in [True, False]:
    for q in [True, False]:
        for r in [True, False]:
            prop1 = not p
            prop2 = not q
            prop3 = (prop1 and prop2) or not(prop1 or prop2)
            prop4 = (q and r) or not (q or r)
            prop5 = (prop3 and prop4) or not (prop3 or prop4)

            tabel.append(pd.DataFrame({
                'P': [p],
                ' ': [' '],
                'Q': [q],
                '  ': [' '],
                'R': [r],
                '   ': [' '],
                '~p': [prop1],
                '    ': [' '],
                '~q': [prop2],
                '      ': [' '],
                '~p <--> ~q': [prop3],
                '       ': [' '],
                'q <--> r': [prop4],
                '        ': [' '],
                '(~p <--> ~q) <--> (q <--> r)': [prop5],
            }))

tabel_soal_1_f = pd.concat(tabel, ignore_index=True)

```

# Jawaban No 2

## a) [¬p ∧ (p ∨ q)] → q

```python
tabel = []
tautologi = True

for p in [True, False]:
    for q in [True, False]:
        prop1 = not p
        prop2 = p or q
        prop3 = prop1 and prop2
        prop4 = not prop3 or q

        if not prop4:
            tautologi = False

        tabel.append(pd.DataFrame({
            'P': [p],
            ' ': [' '],
            'Q': [q],
            '  ': [' '],
            '~P': [prop1],
            '   ': [' '],
            'P || Q': [prop2],
            '    ': [' '],
            '~P && (P || Q)': [prop3],
            '     ': [' '],
            '[~P && (P || Q)] → Q': [prop4]
        }))

tabel_tautologi_2_a = pd.concat(tabel, ignore_index=True)

if tautologi:
    print("Pernyataan adalah tautologi.")
else:
    print("Pernyataan bukan tautologi.")
```

## b) [(p → q) ∧ (q → r)] → (p → r)

```python
tabel = []
tautologi = True

for p in [True, False]:
    for q in [True, False]:
      for r in [True, False]:
        prop1 = not p or q
        prop2 = not q or r
        prop3 = not p or r
        prop4 = prop1 and prop2
        prop5 = not prop4 or prop3

        if not prop5:
            tautologi = False

        tabel.append(pd.DataFrame({
            'P': [p],
            ' ': [' '],
            'Q': [q],
            '  ': [' '],
            'R': [r],
            '  ': [' '],
            'P -> Q': [prop1],
            '   ': [' '],
            'Q -> R': [prop2],
            '    ': [' '],
            'P -> R': [prop3],
            '    ': [' '],
            '(p -> Q) && (Q -> R)': [prop4],
            '     ': [' '],
            '[(p -> Q) && (Q -> R)] -> (P -> R)': [prop5]
        }))

tabel_tautologi_2_b = pd.concat(tabel, ignore_index=True)

if tautologi:
    print("Pernyataan adalah tautologi.")
else:
    print("Pernyataan bukan tautologi.")
```

## c) [p ∧ (p → q)] → q

```python
tabel = []
tautologi = True

for p in [True, False]:
    for q in [True, False]:
        prop1 = not p or q
        prop2 = p and prop1
        prop3 = not prop2 or q

        if not prop3:
            tautologi = False

        tabel.append(pd.DataFrame({
            'P': [p],
            ' ': [' '],
            'Q': [q],
            '  ': [' '],
            'P -> Q': [prop1],
            '   ': [' '],
            'P && (P -> Q)': [prop2],
            '    ': ['  '],
            'P && (P -> Q) -> Q': [prop3],
            '    ': ['   '],
        }))

tabel_tautologi_2_c = pd.concat(tabel, ignore_index=True)

if tautologi:
    print("Pernyataan adalah tautologi.")
else:
    print("Pernyataan bukan tautologi.")
```

## d) [(p ∨ q) ∧ (p → r) ∧ (q → r)] → r

```python
tabel = []
tautologi = True

for p in [True, False]:
    for q in [True, False]:
      for r in [True, False]:
        prop1 = p or q
        prop2 = not p or r
        prop3 = not q or r
        prop4 = prop1 and prop2 and prop3
        prop5 = not prop4 or r

        if not prop5:
            tautologi = False

        tabel.append(pd.DataFrame({
            'P': [p],
            ' ': [' '],
            'Q': [q],
            '  ': [' '],
            'R': [r],
            '  ': [' '],
            'P || Q': [prop1],
            '   ': [' '],
            'Q -> R': [prop2],
            '    ': [' '],
            'P -> R': [prop3],
            '    ': [' '],
            '(p || Q) && (Q -> R) && (P -> R)': [prop4],
            '     ': [' '],
            '[(p || Q) && (Q -> R) && (P -> R)] -> R': [prop5]
        }))

tabel_tautologi_2_d = pd.concat(tabel, ignore_index=True)

if tautologi:
    print("Pernyataan adalah tautologi.")
else:
    print("Pernyataan bukan tautologi.")
```

# Jawaban No 4

Jika G berkata jujur, maka O berkata jujur. (G -> O)

Jika L berkata bohong, maka D berkata jujur. (~L -> D)

Salah satu dari O dan L berkata jujur. (O ^ L)

O berkata jujur jika dan hanya jika D berkata bohong. (O <-> ~D)

```python
tabel = []

for G in [True, False]:
    for O in [True, False]:
        for L in [True, False]:
            for D in [True, False]:
                prop1 = not G
                prop2 = not O
                prop3 = not L
                prop4 = not D
                prop5 = prop1 or O
                prop6 = not prop3 or D
                prop7 = O != L
                prop8 = (O and prop4) or not(O or prop4)
                all_true = prop5 and prop6 and prop7 and prop8

                tabel.append(pd.DataFrame({
                    'G': [G],
                    'O': [O],
                    'L': [L],
                    'D': [D],
                    '~G': [prop1],
                    '~O': [prop2],
                    '~L': [prop3],
                    '~D': [prop4],
                    'G -> O': [prop5],
                    '~L -> D': [prop6],
                    'O ^ L': [prop7],
                    'O <-> ~D': [prop8],
                    'Semua Fakta Benar': [all_true]
                }))

tabel_pembuktian_4 = pd.concat(tabel, ignore_index=True)

if any(tabel_pembuktian_4['Semua Fakta Benar']):
    print("Ada kemungkinan yang benar")
else:
    print("Tidak ada fakta yang benar")
```
